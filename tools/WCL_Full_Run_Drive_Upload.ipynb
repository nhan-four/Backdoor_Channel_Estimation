# WCL full-run package → Google Drive

Notebook này xử lý riêng gói full-run 155.612.278 byte vì OpenCSI raw data và frozen checkpoints đã có trên Drive.

1. Tải `WCL_EXTENSION_FULL_RUN_COMPLETE_20260718.zip` từ cuộc trò chuyện.
2. Chọn **Runtime → Run all**.
3. Cấp quyền đúng tài khoản Google sở hữu thư mục OpenCSI.
4. Khi hộp chọn file xuất hiện, chọn đúng ZIP trên.

Notebook kiểm tra filename, size, SHA-256 và MD5 trước khi upload; sau upload nó đọc lại size và MD5 từ Google Drive. Placeholder cùng tên sẽ được thay byte tại chỗ, nên URL Drive hiện có được giữ ổn định. Chỉ kết thúc khi hiện `TRANSFER FINISHED`.


```python
import pathlib
import requests
import runpy

SCRIPT_URL = (
    'https://raw.githubusercontent.com/nhan-four/'
    'Backdoor_Channel_Estimation/'
    'agent/wcl-full-run-drive-overlay-20260718/'
    'tools/wcl_full_run_drive_upload.py'
)
SCRIPT_PATH = pathlib.Path('/content/wcl_full_run_drive_upload.py')
response = requests.get(SCRIPT_URL, timeout=60)
response.raise_for_status()
SCRIPT_PATH.write_bytes(response.content)
runpy.run_path(str(SCRIPT_PATH), run_name='__main__')
```
